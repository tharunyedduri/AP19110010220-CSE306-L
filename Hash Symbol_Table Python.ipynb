```python
file = open("input.txt", "r")
input = (file.read())
file.close()
#InternalFunctions
def findingIdentifiers(string,datatypes):
    if (string.find("printf") == -1):
        for i in datatypes:
            if(i in string):
                return True
def ReturningIdentifiers(string,datatypes):
    identifiersData=[]
    if(string.find("printf") == -1):
        for i in datatypes:
            if(i in string):
                index = string.index(i);
                indexcolon = string.index(";");
                data = (string[index+len(i):indexcolon].split(","))
                final_ans = []
                for i in data:
                    if(i.find("=")!=-1):
                        final_ans.append(i.split("=")[0])
                    else:
                        final_ans.append((i))
                data = [i.strip() for i in final_ans]
                identifiersData.extend(data)
    return identifiersData

#HashTable
def HashArray(input_List):
    def StringToAscii(input_data):
        String_list = [i for i in str(input_data)]
        ascii = 0
        for i in String_list:
            ascii += ord(i)
        return (ascii)
    HashArray_Input =input_List
    length = len(HashArray_Input)
    HashArray = [[] for i in HashArray_Input]
    ModuloArray = [StringToAscii(i)%length for i in HashArray_Input]
    for i in range(len(ModuloArray)):
        HashArray[ModuloArray[i]].append(HashArray_Input[i])
    return (HashArray)
datatypes = ['int','float','char','double','long','short','unsigned']
keywords = ["int","while","break","for","do","if","float","char","switch","double","short","long","unsigned","sizeof","else","register","extern","static","auto","case","break","volatile","enum","typedef","main"]
input2 = input
input = input.replace("/","/")
if(input.count("/*") % 2 !=0 ):
    print("Comments are not closed")
    input = ""
count_comments = input2.count("/*")
i=0
while(i<count_comments):
    indexstart = input2.index("/*");
    indexend = input2.index("*/");
    input2 = input2[:indexstart-2] + input2[indexend+2:]
    i+=1
input = input2
opt = input2
program_line_by_line = input.split("\n")
program_line_by_line = [i.strip() for i in program_line_by_line]
#ignoring comments
for i in range(len(program_line_by_line)):
    # deleting single line comments
    if(program_line_by_line[i].find("//") != -1):
        program_line_by_line[i] = ""
#creating the array of strings without empty spaces.
new_list=[]
for i in program_line_by_line:
    if(i != ""):
        new_list.append(i)
program_line_by_line = new_list

#variables
identifiers = []
header = []

#program - start
idx = 0
for i in program_line_by_line:
    #header files
    if(i.find("#include")!=-1):
        data = i.split("<")
        header.append(data[1][:-1])
        program_line_by_line[idx] = ""
        i = ""
    #identifiers
    if(findingIdentifiers(i,datatypes)):
        identifiers.extend(ReturningIdentifiers(i,datatypes))
    idx += 1

file = open("Hash_SymbolTable.txt",'w')
file.write("")
file.write(f"""
    {HashArray(identifiers)}
""")
print("Output writtened to output.txt")
file.close()
```

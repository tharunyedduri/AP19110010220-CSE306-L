```python
file = open("input.txt", "r")
input = (file.read())
file.close()


file = open("LinkedList_symboltable.txt", 'w')
file.write("")
file.write(f"""Identifiers:\n""")
file.close()


class Node :
    def __init__(self,data):
        self.data = data
        self.ref = None
        

class LinkedList:
    def __init__(self):
        self.head = None
    def traversal(self):
        if self.head is None:
            print('Linked List is empty')
        else:
            n = self.head
            while n is not None:
                file = open("Linkedsymboltable.txt", 'a')
                file.write("")
                file.write(f"""{n.data}\n""")
                file.close()
                
                n = n.ref
    
        
    def add_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
            n.ref = new_node
            
            

def findingIdentifiers(string, datatypes):
    if (string.find("printf") == -1):
        for i in datatypes:
            if (i in string):
                return True
def ReturningIdentifiers(string, datatypes):
    identifiersData = []
    if (string.find("printf") == -1):
        for i in datatypes:
            if (i in string):
                index = string.index(i);
                indexcolon = string.index(";");
                data = (string[index + len(i):indexcolon].split(","))
                final_ans = []
                for i in data:
                    if (i.find("=") != -1):
                        final_ans.append(i.split("=")[0])
                    else:
                        final_ans.append((i))
                data = [i.strip() for i in final_ans]
                identifiersData.extend(data)
    return identifiersData


input2 = input
input = input.replace("*/", "/*")
if (input.count("/*") % 2 != 0):
    print("Comments are not closed")
    input = ""
count_comments = input2.count("/*")
i = 0
while (i < count_comments):
    indexstart = input2.index("/*");
    indexend = input2.index("*/");
    input2 = input2[:indexstart - 2] + input2[indexend + 2:]
    i += 1
input = input2

lst = input.split("\n")
lst = [i.strip() for i in lst]

for i in range(len(lst)):
    if lst[i].find("//"):
        continue
    lst[i] = ""

datatypes = ['int', 'float', 'char', 'double', 'long', 'short', 'unsigned']

identifiers = []

for i in lst:
    if (findingIdentifiers(i, datatypes)):
        identifiers.extend(ReturningIdentifiers(i, datatypes))
LL = LinkedList()
        
for i in identifiers:
    LL.add_end(i)
    
LL.traversal()
    

```

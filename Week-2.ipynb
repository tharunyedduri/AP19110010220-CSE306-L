```python
file = open("input.txt", "r")
input = (file.read())
file.close()
def findingStringAndList(string, listdata):
    if (string.find("printf") == -1):
        for i in listdata:
            if (string.find(i) != -1):
                return True
def ReturningStringAndList(string, listdata):
    data = []
    for i in listdata:
        if (string.find(i) != -1):
            data.append(i)
    if (len(data) != 0): return data
def findingIdentifiers(string, datatypes):
    if (string.find("printf") == -1):
        for i in datatypes:
            if (i in string):
                return True
def ReturningIdentifiers(string, datatypes):
    identifiersData = []
    if (string.find("printf") == -1):
        for i in datatypes:
            if (i in string):
                index = string.index(i);
                indexcolon = string.index(";");
                data = (string[index + len(i):indexcolon].split(","))
                final_ans = []
                for i in data:
                    if (i.find("=") != -1):
                        final_ans.append(i.split("=")[0])
                    else:
                        final_ans.append((i))
                data = [i.strip() for i in final_ans]
                identifiersData.extend(data)
    return identifiersData
input2 = input
input = input.replace("*/", "/*")
if (input.count("/*") % 2 != 0):
    print("Comments are not closed")
    input = ""
count_comments = input2.count("/*")
i = 0
while (i < count_comments):
    indexstart = input2.index("/*");
    indexend = input2.index("*/");
    input2 = input2[:indexstart - 2] + input2[indexend + 2:]
    i += 1
input = input2
keyword = ["int", "while", "break", "for", "do", "if", "float", "char", "switch", "double", "short", "long", "unsigned",
           "sizeof", "else", "register", "extern", "static", "auto", "case", "break", "volatile", "enum", "typedef",
           'main']
datatypes = ['int', 'float', 'char', 'double', 'long', 'short', 'unsigned']

lst = input.split("\n")
lst = [i.strip() for i in lst]

for i in range(len(lst)):
    if lst[i].find("//"):
        continue
    lst[i] = ""

headers = []

for i in range(len(lst)):
    if lst[i].find("#include"):
        continue
    headers.append(lst[i])
    lst[i] = ""

relational_operators = []
datatypes = []
identifiers = []
keywords = []
for i in lst:
    if "<" in i:
        relational_operators.append('<')
        if '<=' in i:
            relational_operators.append('<=')
            if i.count('<') == 1:
                relational_operators.remove('<')
    if '>' in i:
        relational_operators.append('>')
        if '>=' in i:
            relational_operators.append('>=')
            if i.count('>') == 1:
                relational_operators.remove('>')
    if '==' in i:
        relational_operators.append('==')
    if '!=' in i:
        relational_operators.append('!=')
    if 'int' in i and 'print' not in i:
        datatypes.append("int")
    if 'float' in i:
        datatypes.append("float")
    if 'double' in i:
        datatypes.append("double")

    if 'long' in i:
        datatypes.append("long")

    if 'char' in i:
        datatypes.append("char")

    if 'short' in i:
        datatypes.append("short")

    if (findingStringAndList(i, keyword)):
        keywords.extend(ReturningStringAndList(i, keyword))

    if (findingIdentifiers(i, datatypes)):
        identifiers.extend(ReturningIdentifiers(i, datatypes))

file = open("output.txt", 'w')
file.write("")
file.write(f"""
    identifiers    : {identifiers}
    relational Ops : {relational_operators}
    header         : {headers}
    datatypes      : {datatypes}
    keywords       : {keywords}
""")
print("Output writtened to output.txt")
file.close()
```
